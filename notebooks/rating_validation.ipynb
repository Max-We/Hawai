```python
import pandas as pd

from config import RATINGS_FILE

df = pd.read_csv("../"+RATINGS_FILE)
```

```python
df.head()
```

```python
df["uuid_trunc"] = df["uuid"].apply(lambda x: x[:5])
df["item_title"] = df["item_title"].apply(lambda x: x[:10])
df["rating"] = df["rating"].astype(float)
```

```python
# plot ratings for each user as bar plot
df.groupby("uuid_trunc")["rating"].var().plot.bar()
```

```python
# plot variance ratings for each item ad bar plot
df.groupby("item_title")["rating"].var().plot.bar()
```

```python
# plot rating distribution overall (integer x-axis)
df["rating"].plot.hist(bins=5)
```
